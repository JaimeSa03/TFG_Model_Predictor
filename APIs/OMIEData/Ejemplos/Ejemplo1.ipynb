```python
import datetime as dt
import matplotlib.pyplot as plt

from OMIEData.DataImport.omie_marginalprice_importer import OMIEMarginalPriceFileImporter
from OMIEData.Enums.all_enums import DataTypeInMarginalPriceFile
```

## Pequeño ejemplo de uso de la API

Descargar los datos de el precio por horas y demanda de la energia

*PRICE_SP*  Precio marginal en el mercado español MWh/Euros

*PRICE_PT*  Precio marginal en el mercado portugués MWh/Euros

*ENER_IB_BILLAT* Energía total con bilaterales del mercado iberico MWh

*ENER_IB* Energía total derl mercado Ibérico MWh


```python
dateIni = dt.datetime(2024, 4, 1)
dateEnd = dt.datetime(2024, 4, 8)

# This can take time, it is downloading the files from the website..
df = OMIEMarginalPriceFileImporter(date_ini=dateIni, date_end=dateEnd).read_to_dataframe(verbose=True)
df.sort_values(by='DATE', axis=0, inplace=True)
print(df.head(8))
```

### Limpieza datos 
    Filtramos los datos para obtener solo los de España

```python
# Just prices in spain
str_price_spain = str(DataTypeInMarginalPriceFile.PRICE_SPAIN)
dfPrices = df[df.CONCEPT == str_price_spain]

print(dfPrices.head(10))  # Primeras 50 filas

# Plotting
plt.figure()
plt.plot(dfPrices.DATE, dfPrices.H8, label='Precio luz 8:00')
plt.plot(dfPrices.DATE, dfPrices.H21, label='Precio luz 21:00 pm')
plt.xticks(rotation=45)  # Gira las fechas 45 grados para poder leerlas bien
plt.legend()
plt.show()

```
