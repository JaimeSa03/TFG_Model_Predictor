*Chuleta de Comandos de github

Flujo habitual

```python
git add . 

git commit "Mensaje_commit"

git push origin rama
```

Crear Rama

Recuerda que para que se quede guardada en remoto y no en local, hace falta hacer un push

```python
git branch -m rama_actual rama_nueva

```

Eliminar Rama

```python
git push origin --delete rama
```

Cambiar de Rama

```python
git checkout rama
```

Actualizar main a una rama

```python
# 1. Entrar en main
git checkout main

# 2. Hacer merge de rama# a main
git merge rama#

# 3. Subir los cambios a GitHub
git push origin main
```

Ver las diferencias entre dos ramas

```python
git diff main..jaimes10
```
